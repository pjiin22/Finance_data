```python
# 주가 데이터 로드 (예: 대한항공)
price_data = fdr.DataReader('003490', '2023-01-01', '2023-12-31')  # 대한항공 티커
price_data
```

```python
# 이동 평균 (Moving Average)
price_data["MA20"] = price_data["Close"].rolling(window=20).mean()  # 20일 이동 평균
price_data["MA50"] = price_data["Close"].rolling(window=50).mean()  # 50일 이동 평균

# RSI 계산 (14일 기준)
delta = price_data["Close"].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
price_data["RSI"] = 100 - (100 / (1 + rs))

# MACD 계산
price_data["EMA12"] = price_data["Close"].ewm(span=12, adjust=False).mean()
price_data["EMA26"] = price_data["Close"].ewm(span=26, adjust=False).mean()
price_data["MACD"] = price_data["EMA12"] - price_data["EMA26"]
price_data["MACD_signal"] = price_data["MACD"].ewm(span=9, adjust=False).mean()
price_data["MACD_hist"] = price_data["MACD"] - price_data["MACD_signal"]

# 시각화 예시
plt.figure(figsize=(14, 7))
plt.subplot(3, 1, 1)
plt.plot(price_data.index, price_data["Close"], label="Close Price")
plt.plot(price_data.index, price_data["MA20"], label="20-Day MA")
plt.plot(price_data.index, price_data["MA50"], label="50-Day MA")
plt.legend()
plt.title("주가와 이동평균")

plt.subplot(3, 1, 2)
plt.plot(price_data.index, price_data["RSI"], label="RSI", color='purple')
plt.axhline(70, color='red', linestyle='--')  # 과매수
plt.axhline(30, color='green', linestyle='--')  # 과매도
plt.legend()
plt.title("RSI")

plt.subplot(3, 1, 3)
plt.plot(price_data.index, price_data["MACD"], label="MACD", color='blue')
plt.plot(price_data.index, price_data["MACD_signal"], label="Signal Line", color='red')
plt.bar(price_data.index, price_data["MACD_hist"], color='gray', label="MACD Histogram")
plt.legend()
plt.title("MACD")

plt.tight_layout()
plt.show()
```

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import FinanceDataReader as fdr

# 대한항공과 아시아나 주가 데이터를 각각 로드하고 분기 기준으로 조정
def load_and_prepare_price_data(ticker, start_date, end_date):
    price_data = fdr.DataReader(ticker, start_date, end_date)
    price_data['Date'] = pd.to_datetime(price_data.index)
    price_data['Quarter'] = price_data['Date'].dt.to_period('Q')  # 분기 단위로 조정
    price_quarterly = price_data.groupby('Quarter').last().reset_index()  # 분기별 마지막 날 데이터 선택
    price_quarterly['Date'] = price_quarterly['Quarter'].dt.to_timestamp()  # Date 열 생성
    return price_quarterly[['Date', 'Close']]

# 주가 데이터 로드 (대한항공: '003490', 아시아나: '020560')
korea_air_price = load_and_prepare_price_data('003490', '2023-01-01', '2023-12-31')
asiana_price = load_and_prepare_price_data('020560', '2023-01-01', '2023-12-31')

# Date 열을 datetime 형식으로 통일
combined_financials['Date'] = pd.to_datetime(combined_financials['Date'])
korea_air_price['Date'] = pd.to_datetime(korea_air_price['Date'])
asiana_price['Date'] = pd.to_datetime(asiana_price['Date'])

# 대한항공 재무 데이터와 주가 병합
korea_data = pd.merge(combined_financials, korea_air_price, on='Date', how='inner', suffixes=('', '_대한항공'))
korea_data = korea_data.rename(columns={'Close': '대한항공_Close'})

# 아시아나 재무 데이터와 주가 병합
asiana_data = pd.merge(combined_financials, asiana_price, on='Date', how='inner', suffixes=('', '_아시아나'))
asiana_data = asiana_data.rename(columns={'Close': '아시아나_Close'})

# 대한항공과 아시아나 데이터가 잘 병합되었는지 확인
print("대한항공 데이터 샘플:\n", korea_data.head())
print("아시아나 데이터 샘플:\n", asiana_data.head())

# 상관관계 분석 함수 정의
def plot_correlation_matrix(data, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(title)
    plt.show()

# 대한항공 상관관계 분석
plot_correlation_matrix(
    korea_data[[
        "대한항공_Close", "대한항공_영업이익률", "대한항공_순이익률", "대한항공_매출성장률", "대한항공_유동비율", "대한항공_부채비율"
    ]],
    "대한항공 주가와 주요 재무 지표 간의 상관관계"
)

# 아시아나 상관관계 분석
plot_correlation_matrix(
    asiana_data[[
        "아시아나_Close", "아시아나_영업이익률", "아시아나_순이익률", "아시아나_매출성장률", "아시아나_유동비율", "아시아나_부채비율"
    ]],
    "아시아나 주가와 주요 재무 지표 간의 상관관계"
)

# 회귀 분석 함수 정의
def regression_analysis(data, target, features, company_name):
    X = data[features]
    X = sm.add_constant(X)  # 상수항 추가
    y = data[target]
    model = sm.OLS(y, X).fit()
    print(f"\n{company_name} 회귀 분석 결과")
    print(model.summary())

# 대한항공 회귀 분석
regression_analysis(
    korea_data,
    "대한항공_Close",
    ["대한항공_영업이익률", "대한항공_순이익률", "대한항공_매출성장률"],
    "대한항공"
)

# 아시아나 회귀 분석
regression_analysis(
    asiana_data,
    "아시아나_Close",
    ["아시아나_영업이익률", "아시아나_순이익률", "아시아나_매출성장률"],
    "아시아나"
)
```
